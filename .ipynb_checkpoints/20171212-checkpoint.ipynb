```python
import requests
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
resp = requests.get(
    url,
    cookies={'over18':'1'}
)


print(resp.text)
```

```python
!pip install --user --upgrade beautifulsoup4 lxml html5lib==1.0b8
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(resp.text.encode("utf-8"), "lxml")
div_list = soup.findAll('div',{'class' :'title' })
author_list = soup.findAll('div',{'class' :'author' })
author_list = [a.text.strip() for a in author_list]
print(author_list)
print("\n\n\n")
for div in div_list:
    print(div.find('a').get('href') +"\t\t"+ div.text.strip())
#python list comphression
div_list = [div.find('a').get('href') for div in div_list]
print("\n\n\n")
print(div_list)
```
