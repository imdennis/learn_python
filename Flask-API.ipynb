```python
from flask import Flask, request, make_response, jsonify
from ptt_crawler import PttCrawler

app = Flask('flask-api')


@app.route('/test')
def hello_world():
    message = {'message':'hello world'}
    return jsonify(message)

@app.route('/ptt_crawler', methods=['GET','POST'])
def run_crawler():
    if request.method == 'GET':
        board = request.args.get('board', default = 'Gossiping', type = str)
        page = request.args.get('page', default = 1, type = int)
    
    elif request.method == 'POST':
        board = request.get_json().get('board', 'Gossiping')
        page = request.get_json().get('page', 1)
        
    crawler = PttCrawler('/'+ board +'/', page)
    result = crawler.run()
    return jsonify(result)
```

```python
import json
resp = app.test_client().get('/test')
print(resp.data)
print(json.loads(resp.data.decode()))
```

```python
resp = app.test_client().get('/ptt_crawler?page=2&board=cat')
#print(resp.data)
print(json.loads(resp.data.decode()))
```

```python
resp = app.test_client().post(
    '/ptt_crawler',
    data = json.dumps({'board':'dog', 'page' : 1}),
    content_type='application/json'
)

print(json.loads(resp.data.decode())[0:1])
```
